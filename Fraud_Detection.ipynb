# Fraud Detection

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.metrics import classification_report, confusion_matrix
```

```python
df = pd.read_csv(r'C:\Users\samkh\Downloads\archive (1)\creditcard_2023.csv')
```

```python
print("Dataset Info:")
df.info()
```

```python
print('Tha Data-Base')
df
```

```python
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Class", data=df)
plt.title("Fraud vs Non-Fraud Distribution")
plt.show()
```

```python
fraud_cases = df[df["Class"] == 1]
non_fraud_cases = df[df["Class"] == 0]
print(f"\nFraud Cases: {len(fraud_cases)} | Non-Fraud Cases: {len(non_fraud_cases)}")
print(f"Fraud Percentage: {len(fraud_cases) / len(df) * 100:.4f}%")
```

```python
scaler = StandardScaler()
df["Amount"] = scaler.fit_transform(df["Amount"].values.reshape(-1, 1))
```

```python
df.drop(columns=["id"], inplace=True)
```

```python
X = df.drop(columns=["Class"])
y = df["Class"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=2, stratify=y)

print("\nTraining set size:", X_train.shape)
print("Testing set size:", X_test.shape)
```

## Isolation Forest

```python
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import classification_report, f1_score
```

```python
iso_forest = IsolationForest(n_estimators=100, contamination=0.001, random_state=42)
iso_forest.fit(X_train)
```

```python
y_pred = iso_forest.predict(X_test)
```

```python
y_pred = [1 if val == -1 else 0 for val in y_pred]
```

```python
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import IsolationForest

iso_forest = IsolationForest(
    n_estimators=200,  # More trees improve accuracy
    contamination=0.01,  # Estimated fraud proportion
    max_samples='auto',  # Use all data samples
    random_state=42
)

iso_forest.fit(X_train)
y_pred = iso_forest.predict(X_test)

# Convert predictions (-1 = Fraud, 1 = Non-Fraud) to 0 & 1
y_pred = [1 if i == -1 else 0 for i in y_pred]
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=["Non-Fraud", "Fraud"], yticklabels=["Non-Fraud", "Fraud"])
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()
```

```python
f1 = f1_score(y_test, y_pred)
print(f"F1-Score: {f1:.4f}")
```

```python
lof = LocalOutlierFactor(n_neighbors=10, contamination=0.001)

y_pred_lof = lof.fit_predict(X_test)
y_pred_lof = [1 if val == -1 else 0 for val in y_pred_lof]

conf_matrix_lof = confusion_matrix(y_test, y_pred_lof)

print("\nConfusion Matrix:")
print(conf_matrix_lof)

print("\nClassification Report:")
print(classification_report(y_test, y_pred_lof))
```
